Третье - в столбце `rating` есть редкие значения: EC, RP, K-A, AO.

```python
#посмотрим пример строк с одинаковыми названиями
games.query('name == "Need for Speed: Most Wanted"')
```

Игры с одинаковым названием могут иметь разные годы выхода или платформы. Значит всё в порядке.

Рассмтрим 0 в столбце `year_of_release`.

```python
games.query('year_of_release == 0').head(20)
```

Посмотрим, имеют ли эти игры дубликаты с датами.

```python
games.query('name == "Hitman 2: Silent Assassin"')
```

```python
#создадим функцию, заменяющую 0 на медиальные значения даты для строк с дубликатами
def filling_date(data):
    filling_year = []
    if data['year_of_release'] == 0:
        name_dup = data['name']
        data['year_of_release'] = games.query('(name == @name_dup) & (year_of_release != 0)')['year_of_release'].median()
        return data['year_of_release']

#применим функцию     
games.loc[games['year_of_release'] == 0, 'year_of_release'] = games.apply(filling_date, axis=1)

#проверим, что нулей не осталось
display(games.query('year_of_release == 0').head(20))

#убедимся, что 0, которые не на что заменить удалились
games.isna().sum()
```

Некоторые медиальные значения стали дробными числами.

```python
games.query('name == "Hitman 2: Silent Assassin"')
```

Строки без указания даты релиза для дальнейшего анализа не имеют смысла. Удалим их и повторно преобразуем тип данных в столбце с годом.

```python
games = games.dropna(subset=['year_of_release'])
games.isna().sum()
```

```python
#преобразуем в uint
games['year_of_release'] = np.uint16(games['year_of_release'])
games.info()
```

Перейдем к столбцу `rating`, в котором есть редкие значения: EC, RP, K-A, AO.

Согласно информации о ESRB:

* EC - для детей от 3 лет.
* RP - ожидает присвоения рейтинга.
* K-A - от 6 лет (устар.).
* AO - старше 18 лет.

Распространенные в таблице маркировки имеют значения:
* E - от 6 лет.
* T - от 13 лет
* M - от 17 лет.
* E10+ - от 10 лет.

Таким образом мы можем объединить K-A и E. Значение RP заменить на NaN, так как оно не определено. А строки с EC и AO удалить как редкие и выбивающиеся значения.

```python
games['rating'] = games['rating'].replace('K-A', 'E')
games['rating'] = games['rating'].replace('RP', np.NaN)
```

```python
games = games.query('~(rating == "EC") & ~(rating == "AO")')

#проверим результат
games.info()
```

Успешно.

## Проверка гипотез

Проверим гипотезы:
1. Гипотеза № 1: 
    * Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    * Альтернативаня гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные;
2. Гипотеза № 2: 
    * Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.
    * Альтернативаня гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

Воспользуемся тестом Стьюдента для проверки гипотез о равенстве средних двух выборок, т.к. выборки удовлетворяют условиям:
* Генеральные совокупности не зависят друг от друга.
* Выборочные средние нормально распределены, т.к. в выборках больше 100 значений.

Проверим Гипотезу № 1 по данным таблицы `games_actual`.

```python
#выборка пользовательского рейтинга платформы XOne
df_xone_rating = games_actual.query('platform == "XOne"')['user_score']

#выборка пользовательского рейтинга платформы PS4
df_pc_rating = games_actual.query('platform == "PC"')['user_score']
```

```python
#посмотрим параметры распределений в выборках
print('Параметры распределения пользовательского рейтинга платформы XOne:')
display(df_xone_rating.describe())
print('Параметры распределения пользовательского рейтинга платформы PC:')
display(df_pc_rating.describe())
```

По предворительным данным видим, что выборки очень пожожи.

```python
#построим гистогрммы распределения

df_xone_rating.hist(bins=10, alpha=0.7);
df_pc_rating.hist(bins=10, alpha=0.7);
```

Выборки имеют схожий вид.

Применим тест Стьюдента.

```python
#выполним проверку на равенство средних, установим уровень статистической значимости 0.05
alpha = 0.05
results_rating_pl = st.ttest_ind(df_xone_rating.dropna(), df_pc_rating.dropna())
print('p-значение:', results_rating_pl.pvalue)
if results_rating_pl.pvalue < alpha:
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативы. Средние пользовательские рейтинги платформ Xbox One и PC вероятно разные.")
else:
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) нет оснований отвергнуть нулевую гипотезу. Средние пользовательские рейтинги платформ Xbox One и PC вероятно равны.")
```

Проверим Гипотезу № 2.

```python
#выборка пользовательского рейтинга жанра Action
df_action_rating = games_actual.query('genre == "Action"')['user_score']

#выборка пользовательского рейтинга жанра Shooter
df_sports_rating = games_actual.query('genre == "Sports"')['user_score']
```

```python
#посмотрим параметры распределений в выборках
print('Параметры распределения пользовательского рейтинга жанра Action:')
display(df_action_rating.describe())
print('Параметры распределения пользовательского рейтинга жанра Sports:')
display(df_sports_rating.describe())
```

По предворительным данным видим, что выборки достаточно сильно отличаются, в том числе по размеру.

```python
#построим гистогрммы распределения

df_action_rating.hist(bins=10, alpha=0.6);
df_sports_rating.hist(bins=10, alpha=0.8);
```

Выборки отличаются не только по высоте (может быть обусловлено разницей в размерах выборок), но и расположению пиков.

Применим тест Стьюдента.

```python
#выполним проверку на равенство средних, оставим уровень статистической значимости 0.05
results_rating_genre = st.ttest_ind(df_action_rating.dropna(), df_sports_rating.dropna())
print('p-значение:', results_rating_genre.pvalue)
if results_rating_genre.pvalue < alpha:
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативы. Средние пользовательские рейтинги жанров Action и Sports вероятно разные.")
else:
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) нет оснований отвергнуть нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports вероятно равны.")
```

В таком случае выдвиним и проверим дополнительную гипотезу.

Гипотеза № 3:
* Нулевая гипотеза: средний пользовательский рейтинг жанра Action выше, чем Sports.
* Альтернативаня гипотеза: средний пользовательский рейтинг жанра Action ниже, чем Sports.

```python
if (results_rating_genre.pvalue / 2 < alpha) and (df_action_rating.mean() <= df_sports_rating.mean()):
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) есть основания отвергнуть нулевую гипотезу в пользу альтернативы. Средний пользовательский рейтинг жанра Action вероятно ниже, чем Sports.")
else:
    print("На имеющихся данных, на уровне значимости 5% (уровне доверия 95%) нет оснований отвергнуть нулевую гипотезу. Средний пользовательский рейтинг жанра Action вероятно выше, чем Sports.")
```

Вывод подтверждает данные, полученные нами в п. 3.8.
