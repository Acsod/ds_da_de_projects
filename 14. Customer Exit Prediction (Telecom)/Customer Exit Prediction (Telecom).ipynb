# Прогнозирование оттока клиентов


**Иcходные данные**

Оператор связи хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. 

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Цель исследования**

Нам необходимо построить модель для задачи классификации, способную с высокой вероятностью предсказать, собирается ли клиент отказаться от услуг компании.

**Ход исследования**
1. Обзор и предобработка данных.
2. Исследовательский анализ данных.
3. Обучение и проверка различных моделей (LightGBM, CatBoost, RandomForest, LogisticRegression (ElasticNet)).
4. Провека качества лучшей модели на тестовой выборке.

### Предобрабтка данных

Для начала нам необходимо преодобработать данные. Из предобзора видим, что явных пропусков и аномальных значени не обнаружено.
1. В таблице contract создать целевой признак из столбца end_data. Посчитать количество дней от начала 1ого контракта. Удалить столбцы begin_date и end_date.
2. Проверить данные на дубликаты.
3. Соединить таблицы по признаку customer_id.

```python
# создадим целевой признак

def target(date):
    if date == 'No':
        data = 0
    else:
        data = 1
    return data

contract['exited'] = contract['end_date'].apply(target)
```

```python
# изменим в столбце end_date значения 'No' на дату выгрузки

contract.loc[contract['end_date'] == 'No', 'end_date'] = '2020-02-01 00:00:00'
contract['end_date'] = pd.to_datetime(contract['end_date'])
contract['begin_date'] = pd.to_datetime(contract['begin_date'])

# создадим столбец days_with_company

contract['days_with_company'] = contract['end_date'] - contract['begin_date']
contract['days_with_company'] = np.uint16(contract['days_with_company'].dt.days)

# проверим результат

contract['days_with_company']
```

```python
# удалим столбцы begin_date и end_date, чтобы избежать утечки данных

contract = contract.drop(['end_date', 'begin_date'], axis=1)
df_list[0] = contract

# посмотрим на таблицу

contract.head()
```

В остальных таблицах, согласно обзору данных, преобразования не требуются. Но предворительно проверим все таблицы на дублекаты.

Также отдельно проверим дублекаты среди ID клиентов. Если они уникальны, можно перевести их в индексы и упростить объединение таблиц.

```python
for i in range(len(df_list)):
    print(f'Дублекатов в таблице: {list_tables[i]}')
    print(df_list[i].duplicated().sum())
    print('Дублекатов customer_id:')
    print(df_list[i]['customer_id'].duplicated().sum())
    print()
```

Дублекаты отсутствуют. 

Заменим индексы на customer_id и объединим таблицы.

```python
for i in range(len(df_list)):
    df_list[i] = df_list[i].set_index('customer_id')
```

```python
# объединим таблицы
df = pd.concat(df_list, axis=1, join='outer')

print(df.info())
df.head()
```

Объединение выполнено успешно. Однако у нас образовались пропуски, которые необходимо заполнить.

Идентифицировать наличие услуги internet у пользователя мы сможем по столбцу internet_service, заполнив пропуски значением No. Пропуски в остальных столбцах также можно заполнить значением No.

С таблицей phone несколько сложнее. Мы можем заполнить пропуски в multiple_lines каким-либо 3 значением или добавить еще один столбец, позволяющий идентифицировать наличие мобильно связи у пользователя, а пропуски заполнить No. Второй вариант кажется логичнее.

```python
# добавим столбец для идентификации услуги телефонии
def is_phone(value):
    if value == 'No' or value == 'Yes':
        return 1
    else:
        return 0

df['is_phone'] = df['multiple_lines'].apply(is_phone)
```

```python
# заполним пропуски

df = df.fillna('No')

# помотрим итоговую таблицу
print(df.info())
df.head(10)
```

## Обучение и проверка различных моделей

### Выделение целевого признака

```python
target = df['exited']
features = df.drop(['exited'], axis=1)
```

### Разделение данных на выборки

```python
features_train, features_test, target_train, target_test = train_test_split(features, 
                                                                            target, 
                                                                            test_size=0.25, 
                                                                            random_state=141122, 
                                                                            stratify=target)
```

Проверим соотношение.

```python
print(f'{features_test.shape[0] / features.shape[0]:.2%}')
```

### Масштабирование данных

В таблице присутствуют признаки с разными разбросами значений. Приведем их к общему масштабу для правильной работы линейной модели.

```python
# выделим признаки для масштабирования

scaler_features = ['monthly_charges', 'days_with_company']

# создадим и обучим объект для масштабирования

scaler = StandardScaler()
scaler.fit(features_train[scaler_features])

# выделим в отдельные переменные

features_train_sc = features_train.copy()
features_test_sc = features_test.copy()

# масштабируем

fit_sc = pd.DataFrame(scaler.transform(features_train[scaler_features]))
test_sc = pd.DataFrame(scaler.transform(features_test[scaler_features]))
fit_sc.index = features_train.index
test_sc.index = features_test.index

features_train_sc[scaler_features] = fit_sc
features_test_sc[scaler_features] = test_sc

# проверим результат
pd.options.mode.chained_assignment = None
features_train_sc.head()
```

### Кодирование данных

Для линейной модели и случайного леса данные необходимо закодировать.

Закодируем отмасштабированные данные.

```python
# выделим признаки для кодирования

ohe_features = features_train_sc.columns.drop(scaler_features)

# закодируем данные

ohe = OneHotEncoder(drop='first')
ohe.fit(features_train_sc[ohe_features])
new_columns = ohe.get_feature_names(ohe_features)

# тренировочная выборка

features_train_sc_ohe = pd.DataFrame(ohe.transform(features_train_sc[ohe_features]).toarray(), columns=new_columns)
features_train_sc_ohe.index = features_train_sc.index
features_train_sc_ohe[scaler_features] = features_train_sc[scaler_features]

# тестовая выборка

features_test_sc_ohe = pd.DataFrame(ohe.transform(features_test_sc[ohe_features]).toarray(), columns=new_columns)
features_test_sc_ohe.index = features_test_sc.index
features_test_sc_ohe[scaler_features] = features_test_sc[scaler_features]

# проверим результат
features_train_sc_ohe.head()
```

Закодируем неотмасштабированные данные.

```python
# тренировочная выборка

features_train_ohe = features_train_sc_ohe
features_train_ohe[scaler_features] = features_train[scaler_features]

# тестовая выборка

features_test_ohe = features_test_sc_ohe
features_test_ohe[scaler_features] = features_test[scaler_features]

# проверим результат
features_train_ohe.head()
```

Мы выполнили масштабирование и кодирование. Можем приступать к обучению моделей.

### Обучение моделей

Проверим следующие модели:

1. LightGBM
2. CatBoost
3. RandomForest
4. LogisticRegression (ElasticNet)

#### LightGBM

Обучим LightGBM на немасштабированных данных.

```python
%%time
# 6-7 сек

param_lgbm = {'learning_rate': np.arange(0.1, 1.0, 0.1),
              'n_estimators': range(10, 150, 30),
              'reg_alpha': [0.2],
              'reg_lambda': [0.1]}

lgbm_gs = GridSearchCV(LGBMClassifier(random_state=141122,
                                      class_weight='balanced'), 
                       param_lgbm, 
                       scoring='roc_auc', 
                       n_jobs=-1).fit(features_train_ohe, target_train)

model_lgbm = lgbm_gs.best_estimator_
score_lgbm = lgbm_gs.best_score_
fit_time_lgbm = np.mean(lgbm_gs.cv_results_['mean_fit_time'])
score_time_lgbm = np.mean(lgbm_gs.cv_results_['mean_score_time'])

print(f'ROC AUC LightGBM на валидационной выборке: {score_lgbm:.2f}')
print(f'Время обучения LightGBM: {fit_time_lgbm:.2f} c')
print(f'Время предсказания LightGBM: {score_time_lgbm:.2f} c')
```

#### CatBoost

Обучим CatBoost на немасштабированных данных.

```python
%%time
# около 5 минут

param_cb = {'depth': range(2, 16, 2),
            'n_estimators': range(10, 150, 30),
            'learning_rate': [0.5],
            'reg_lambda': [0.1]}

cb_gs = GridSearchCV(CatBoostClassifier(random_state=141122,
                                        verbose=False), 
                      param_cb, 
                      scoring='roc_auc', 
                      n_jobs=-1).fit(features_train_ohe,
                                     target_train)

model_cb = cb_gs.best_estimator_
score_cb = cb_gs.best_score_
fit_time_cb = np.mean(cb_gs.cv_results_['mean_fit_time'])
score_time_cb = np.mean(cb_gs.cv_results_['mean_score_time'])

print(f'ROC AUC CatBoost на валидационной выборке: {score_cb:.2f}')
print(f'Время обучения CatBoost: {fit_time_cb:.2f} c')
print(f'Время предсказания CatBoost: {score_time_cb:.2f} c')
```

#### RandomForest

Обучим RandomForest на немасштабированных данных.

```python
%%time
# 6 сек

param_rf = {'max_depth': range(2, 41, 2),
            'n_estimators': range(10, 201, 30)}

rf_rs = RandomizedSearchCV(RandomForestClassifier(random_state=141122,
                                                  class_weight='balanced'), 
                       param_rf, 
                       scoring='roc_auc', 
                       n_jobs=-1).fit(features_train_ohe, target_train)

model_rf = rf_rs.best_estimator_
score_rf = rf_rs.best_score_
fit_time_rf = np.mean(rf_rs.cv_results_['mean_fit_time'])
score_time_rf = np.mean(rf_rs.cv_results_['mean_score_time'])

print(f'ROC AUC RandomForest на валидационной выборке: {score_rf:.2f}')
print(f'Время обучения RandomForest: {fit_time_rf:.2f} c')
print(f'Время предсказания RandomForest: {score_time_rf:.2f} c')
```

```python
feature_importance = pd.Series(model_rf.feature_importances_,
                               features_train_ohe.columns).sort_values(ascending=False)
feature_importance
```

#### LogisticRegression (ElasticNet)

Обучим LogisticRegression на масштабированных данных.

```python
%%time

# 3 мин

param_en = {'l1_ratio': np.arange(0.1, 1.1, 0.1),
            'max_iter': range(20, 500, 30)}

en_gs = GridSearchCV(LogisticRegression(random_state=141122,
                                        solver='saga',
                                        penalty='elasticnet',
                                        class_weight='balanced'), 
                       param_en, 
                       scoring='roc_auc', 
                       n_jobs=-1).fit(features_train_sc_ohe, target_train)

model_en = en_gs.best_estimator_
score_en = en_gs.best_score_
fit_time_en = np.mean(en_gs.cv_results_['mean_fit_time'])
score_time_en = np.mean(en_gs.cv_results_['mean_score_time'])

print(f'ROC AUC LogisticRegression (ElasticNet) на валидационной выборке: {score_en:.2f}')
print(f'Время обучения LogisticRegression (ElasticNet): {fit_time_en:.2f} c')
print(f'Время предсказания LogisticRegression (ElasticNet): {score_time_en:.2f} c')
```
